```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
```

```python
x = np.linspace(-1,1,10001)
y = norm.pdf(x)
plt.plot(x,y)
```

```python
np.sum(y * (x[1]-x[0]))
```

```python
norm.cdf(1)-norm.cdf(-1)
```
