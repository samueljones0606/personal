```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax =plt.subplots(1,1)

for i in range(5):
    x = pd.Series(np.random.uniform(size=100)).rank().rename("x")
    y = pd.Series(np.random.uniform(size=100)).rank().rename("y")
    df = pd.concat((x,y), axis=1)
    theta = np.linspace(0,1,11)
    theta =theta.reshape(1,-1)
    pd.DataFrame(df["x"].values.reshape(-1,1)*theta + df["y"].values.reshape(-1,1)*(1-theta), columns=theta.flatten()).rank().corr("spearman").iloc[0].plot(ax=ax)
```
